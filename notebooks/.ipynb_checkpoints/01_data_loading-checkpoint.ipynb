```python
def generate_rul(df):
    rul = df.groupby('unit_number')['time_in_cycles'].max().reset_index()
    rul.columns = ['unit_number', 'max_cycle']
    df = df.merge(rul, on='unit_number', how='left')
    df['RUL'] = df['max_cycle'] - df['time_in_cycles']
    return df

train_FD001 = generate_rul(train_FD001)
print(train_FD001[['unit_number','time_in_cycles','RUL']].head())
```

```python
# from scipy.stats import skew, entropy

# def extract_features_from_file(file_path):
#     df = pd.read_csv(file_path, header=None)
#     df.columns = ["Hour","Minute","Second","Microsecond","Horiz_accel","Vert_accel"]

#     feats = {}
#     for axis in ["Horiz_accel","Vert_accel"]:
#         signal = df[axis].values

#         # Statistical features
#         feats[f"RMS_{axis}"] = np.sqrt(np.mean(signal**2))
#         feats[f"Kurtosis_{axis}"] = pd.Series(signal).kurt()
#         feats[f"Skew_{axis}"] = skew(signal)
#         feats[f"Entropy_{axis}"] = entropy(np.abs(signal)+1e-10)  # small offset to avoid log(0)
#         feats[f"CrestFactor_{axis}"] = np.max(np.abs(signal)) / (np.sqrt(np.mean(signal**2)) + 1e-8)

#         # FFT energy bands
#         energies = fft_band_energy(signal)
#         for i,e in enumerate(energies):
#             feats[f"FFTband{i+1}_{axis}"] = e

#     return feats



# from scipy.fft import fft

# def fft_band_energy(signal, fs=25600, bands=[(0,1000),(1000,5000),(5000,10000)]):
#     N = len(signal)
#     freqs = np.fft.fftfreq(N, d=1/fs)
#     fft_vals = np.abs(fft(signal))**2
#     energies = []
#     for (low,high) in bands:
#         mask = (freqs >= low) & (freqs < high)
#         energies.append(fft_vals[mask].sum())
#     return energies




from scipy.stats import skew, entropy
from scipy.fft import fft
import pywt
import pandas as pd
import numpy as np

def extract_features_from_file(file_path):
    df = pd.read_csv(file_path, header=None)
    df.columns = ["Hour","Minute","Second","Microsecond","Horiz_accel","Vert_accel"]

    feats = {}
    for axis in ["Horiz_accel","Vert_accel"]:
        signal = df[axis].values

        # Statistical features
        feats[f"RMS_{axis}"] = np.sqrt(np.mean(signal**2))
        feats[f"Kurtosis_{axis}"] = pd.Series(signal).kurt()
        feats[f"Skew_{axis}"] = skew(signal)
        feats[f"Entropy_{axis}"] = entropy(np.abs(signal)+1e-10)  # avoid log(0)
        feats[f"CrestFactor_{axis}"] = np.max(np.abs(signal)) / (np.sqrt(np.mean(signal**2)) + 1e-8)

        # FFT energy bands
        energies = fft_band_energy(signal)
        for i, e in enumerate(energies):
            feats[f"FFTband{i+1}_{axis}"] = e

        # Wavelet transform features
        coeffs = pywt.wavedec(signal, 'db4', level=3)
        for i, c in enumerate(coeffs):
            feats[f"WaveletEnergy_L{i}_{axis}"] = np.sum(np.square(c))

    return feats


def fft_band_energy(signal, fs=25600, bands=[(0,1000),(1000,5000),(5000,10000)]):
    N = len(signal)
    freqs = np.fft.fftfreq(N, d=1/fs)
    fft_vals = np.abs(fft(signal))**2
    energies = []
    for (low, high) in bands:
        mask = (freqs >= low) & (freqs < high)
        energies.append(fft_vals[mask].sum())
    return energies
```

```python
base_dir = "../data/raw/PRONOSTIA/Learning_set/"

all_records = []
for bearing_folder in os.listdir(base_dir):
    folder_path = os.path.join(base_dir, bearing_folder)
    if os.path.isdir(folder_path):  # only process folders
        for fname in sorted(os.listdir(folder_path)):
            if fname.startswith("acc_") and fname.endswith(".csv"):
                fpath = os.path.join(folder_path, fname)
                feats = extract_features_from_file(fpath)
                feats["file"] = fname
                feats["bearing_id"] = bearing_folder
                all_records.append(feats)

features_df = pd.DataFrame(all_records)
print(features_df.head())
print(features_df["bearing_id"].unique())
```

```python
# --- Improved failure labeling based on vibration threshold ---
FAIL_THRESHOLD = 20.0  # as per PHM 2012 challenge description

def assign_binary_labels(features_df):
    """
    Assign binary failure labels to vibration features.
    Label = 1 (failing) if RMS in either axis exceeds threshold, else 0 (healthy).
    """
    features_df["label"] = (
        (features_df["RMS_Horiz_accel"] > FAIL_THRESHOLD) |
        (features_df["RMS_Vert_accel"] > FAIL_THRESHOLD)
    ).astype(int)
    return features_df

# Example usage:

features_df = assign_binary_labels(features_df)
print(features_df["label"].value_counts())
features_df = features_df.rolling(window=5, min_periods=1).mean()
```

```python
def process_bearing_folder(bearing_dir, out_path):
    records = []
    for fname in sorted(os.listdir(bearing_dir)):
        if fname.startswith("acc_") and fname.endswith(".csv"):
            fpath = os.path.join(bearing_dir, fname)
            feats = extract_features_from_file(fpath)
            feats["file"] = fname
            records.append(feats)
    df = pd.DataFrame(records)
    max_idx = len(df)
    df["RUL"] = list(range(max_idx-1, -1, -1))
    df["label"] = 0
    df.loc[df["RUL"] < max_idx*0.2, "label"] = 1
    df.to_csv(out_path, index=False)
    print(f"Processed and saved {out_path}")
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

sensor_cols = [c for c in train_FD001.columns if "sensor" in c]

def add_lag_features(df, sensor_cols, lags=[5, 10]):
    for col in sensor_cols:
        for lag in lags:
            df[f"{col}_lag{lag}"] = df.groupby('unit_number')[col].shift(lag)
    df = df.dropna()  # drop rows with NaN due to lagging
    return df
train_FD001 = add_lag_features(train_FD001, sensor_cols)



scaler = StandardScaler()
X_scaled = scaler.fit_transform(train_FD001[sensor_cols])

pca = PCA(n_components=1)
train_FD001["health_index"] = pca.fit_transform(X_scaled)

# Plot health index for a sample engine
sample_engine = train_FD001[train_FD001["unit_number"]==1]
plt.plot(sample_engine["time_in_cycles"], sample_engine["health_index"])
plt.gca().invert_yaxis()
plt.title("Health Index (PCA) - Engine 1")
plt.xlabel("Cycle")
plt.ylabel("Health Index")
plt.show()

train_FD001 = add_lag_features(train_FD001, sensor_cols)

```

```python
import os
import pandas as pd
from sklearn.model_selection import GroupShuffleSplit
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

# Paths
base_dir = "../data/raw/PRONOSTIA/Learning_set/"
out_dir = "../data/processed/"

# --- Merge all processed bearing CSVs ---
all_records = []
for bearing_folder in os.listdir(base_dir):
    out_path = os.path.join(out_dir, f"{bearing_folder}.csv")
    if os.path.exists(out_path):
        df = pd.read_csv(out_path)
        df["bearing_id"] = bearing_folder  # tag each row
        all_records.append(df)

features_df = pd.concat(all_records, ignore_index=True)

print("Bearings included:", features_df["bearing_id"].unique())
print("Shape:", features_df.shape)
print(features_df.head())

# --- Grouped train/test split ---
X = features_df.drop(columns=['label', 'file', 'bearing_id'], errors='ignore')
y = features_df['label']
groups = features_df['bearing_id']  # ensure no leakage

gss = GroupShuffleSplit(test_size=0.2, n_splits=1, random_state=42)
train_idx, test_idx = next(gss.split(X, y, groups))

X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

# --- Train Logistic Regression ---
log_reg = LogisticRegression(max_iter=1000)
log_reg.fit(X_train, y_train)

y_pred = log_reg.predict(X_test)
print("\nPRONOSTIA Logistic Regression (Grouped Split)")
print(classification_report(y_test, y_pred))
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression

pipe = Pipeline([
    ("scaler", StandardScaler()),
    ("log_reg", LogisticRegression(max_iter=5000, solver="lbfgs"))
])

pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
from sklearn.svm import OneClassSVM
from sklearn.ensemble import IsolationForest
from sklearn.metrics import classification_report

# Train One-Class SVM only on healthy bearings
X_healthy = X_train[y_train==0]
oc_svm = OneClassSVM(kernel="rbf", gamma="scale", nu=0.05)
oc_svm.fit(X_healthy)

y_pred_svm = oc_svm.predict(X_test)
y_pred_svm = (y_pred_svm == -1).astype(int)  # -1 = anomaly â†’ 1 (fail)

print("One-Class SVM")
print(classification_report(y_test, y_pred_svm))

# Train Isolation Forest
iso_forest = IsolationForest(contamination=0.1, random_state=42)
iso_forest.fit(X_healthy)

y_pred_iso = iso_forest.predict(X_test)
y_pred_iso = (y_pred_iso == -1).astype(int)

print("Isolation Forest")
print(classification_report(y_test, y_pred_iso))
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

def create_sequences(data, target, window=30):
    X, y = [], []
    for i in range(len(data)-window):
        X.append(data[i:i+window])
        y.append(target[i+window])
    return np.array(X), np.array(y)

# Example on one unit (unit 1)
unit1 = train_FD001[train_FD001['unit_number']==1]
X_seq, y_seq = create_sequences(unit1.drop(columns=['unit_number','time_in_cycles','RUL']).values,
                                unit1['RUL'].values)

# Train/val split
split = int(0.8*len(X_seq))
X_train, X_val = X_seq[:split], X_seq[split:]
y_train, y_val = y_seq[:split], y_seq[split:]

# Build LSTM
model = Sequential([
    LSTM(64, input_shape=(X_train.shape[1], X_train.shape[2])),
    Dense(1)
])
model.compile(optimizer="adam", loss="mse", metrics=["mae"])
history = model.fit(X_train, y_train, validation_data=(X_val, y_val),
                    epochs=20, batch_size=32, verbose=1)
```
